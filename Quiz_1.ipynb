```python
georgians = df[(df['Country'] == 'Georgia')] 
georgian_windows_users = georgians['OpSysPersonal use'].str.contains('Windows', na=False).sum()
georgian_macos_users = georgians['OpSysPersonal use'].str.contains('MacOS', na=False).sum() 
georgian_ubuntu_users = georgians['OpSysPersonal use'].str.contains('Ubuntu', na=False).sum()
```

```python
import matplotlib.pyplot as plt                    #ქართველი მომხმარებლების ოპერაციული სისტემების გადანაწილება

fig, ax = plt.subplots()

os = ['Windows', 'MacOS', 'Ubuntu']
user_count = [georgian_windows_users, 
              georgian_macos_users, 
              georgian_ubuntu_users]  

bar_labels = ['Windows', 'MacOS', 'Ubuntu']
bar_colors = ['tab:blue', 'tab:green', 'tab:red']

ax.bar(os, user_count, label=bar_labels, color=bar_colors)

ax.set_ylabel('მომხმარებლების რაოდენობა',weight='bold',size=14)
ax.set_xlabel('ოპერაციული სისტემა',weight='bold',size=14)
ax.set_title('ქართველი მომხმარებლების ოპერაციული სისტემები',weight='bold',size=14)
ax.legend(title='OS')

plt.show()
```

```python
countries = df['Country']

country_names = countries.value_counts().index.tolist() #ქვეყნების დასახელებები ლისტში
user_counts = countries.value_counts().values.tolist()  #ქვეყნების მომხმარებლები ლისტში

small_countries = 0
for count in user_counts[6:]:  #ვაჯამებ ქვეყნებს რომლებიც არ არიან უდიდეს 6-ს შორის
    small_countries += count   #რათა გრაფაში 'other countries'-ს სახით გამოვიტანო
small_countries
user_counts = user_counts[:6]
user_counts.append(small_countries) 

country_names = country_names[:6]       #პატარა ქვეყნებს წარმოვადგენ ერთდ
country_names.append("Other countries")
country_names[3] = "Uk and Northern Ireland"   #ვამოკლებ სახელს
```

```python
import matplotlib.pyplot as plt

labels = country_names
sizes = user_counts

fig, ax = plt.subplots()
ax.axis('equal')
plt.title('მომხმარებლების გადანაწილება ქვეყნების მიხედვით', size=18, weight='bold')

wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0, wedgeprops={'edgecolor': 'black', 'linewidth': 1})

# Customize the appearance of the labels
for text, autotext in zip(texts, autotexts):
    text.set(size=10, weight='bold')
    autotext.set(size=10)


print()
```
